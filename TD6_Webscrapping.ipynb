### Simplification d'écriture pour trouver les éléments web

Lors du webscrapping, j'aurai à rechercher des elements dans les pages web par leur selecteur css ou encore par xpath. Quoiqu'il en soit, je me fais une petite fonction pour qu'en une ligne on puisse traiter tous les cas.
Cela me permets aussi de gérer les erreurs et de retourner un None lorsqu'il y a une erreur où quand ma recherche n'a pas abouti

```python
def find_element(driver, by, pattern, debug=False):
    element = None
    if by == 'css_selector':
        try: element = driver.find_element_by_css_selector(pattern)
        except: pass
    elif by == 'xpath':
        try: element = driver.find_element_by_xpath(pattern)
        except: pass
    
    if element is None and debug:
        print(f'The element was not found by {by} "{pattern}" in the driver {driver}.')
    return element
```

Aussi, pour que le webscrapping soit visuellement plus agréable et plus ressemblant à la navigatio humaine sur internet, je me fais une petite fonction sleep qui permet d'attendre un certain temps avec de l'aléatoire. 
Cela simulera entre autres, l'écriture des variables dans les champs de filtre pour la recherche.

```python
def sleep(t=0.15, minimum=0.05, std_ratio=1):
    time.sleep(max(minimum, np.random.normal(t, t * std_ratio)))
```

#### Trouver les items

J'aimerai avoir une logique d'itération "tant qu'il y a des items", les extraires et rester sur la même page. Alors je me fais une petite fonction pour trouver tous les items et préciser si ce sont des container (pubs) ou non.

#### Extraires les informations des items

Enfin, pour les items qui sont bien des annonces il faudra bien évidemment les extraires. Alors analysons de quoi ils sont composés.

![Image données des annonces](src/item.png)

On peut alors constater que les items peuvent être constituer des données suivantes :
- flag d'exclusivité **[bool]**
- nombre de photo **[int]**
- un ou plusieurs types de bien dont on précisera la liste **[bool]s**
- surface du bien **[int]**
- nombre de pièce **[int]**
- nombre de chambre **[int]**
- lieu et code postal **[str]** et **[str]**
- prix d'achat **[int]**
- publicité pour prix de location mensuel avec partenaire **[int]**

Si un champ n'est pas trouvable, on considéreras que celui-ci n'est pas existant, il ne sera alors pas précisé dans le dataframe final (**NaN**) sauf pour le nombre de photo qu'on considéras être **1** par défaut



```python
types = ['vente', 'appartement', 'maison', 'terrain', 'villa', 'viager', 'propriété']

def extract_data(item):
    data = {}
    soup = BeautifulSoup(item.get_attribute('innerHTML'))
    
    data["flag"] = soup.find('div', {"class": 'flag-container'})
    data["flag"] = data["flag"] is not None and data["flag"].get_text() != ''
    
    # find and extract the number of picture if exists else 1
    try: 
        n_pict_text = soup.find('span', {"class": 'offer-picture-count'}).get_text()
        n_pict_text = n_pict_text.replace('\n', ' ')
        data["n_picture"] = int(re.sub(r'.*\d+/(\d+).*', r'\g<1>', n_pict_text))
    except: data["n_picture"] = 1
    
    # extract the data types separated by a space in the items
    data_types = soup.find('span', {"class": 'offer-details-type'})    
    data_types = data_types.get_text().lower().split(' ') if data_types is not None else []
    for t in types:
        data[f'is_{t}'] = (t in data_types) * 1
        
    # extract the area and the number of room/bedroom 
    data['area'] = soup.find('span', {"class": 'offer-area-number'})
    data['area'] = int(data['area'].get_text()) if data['area'] is not None else None
    data['n_room'] = soup.select('.offer-details-caracteristik--rooms > span')
    data['n_room'] = int(data['n_room'][0].get_text()) if len(data['n_room']) else None
    data['n_bedroom'] = soup.select('.offer-details-caracteristik--bedrooms > span')
    data['n_bedroom'] = int(data['n_bedroom'][0].get_text()) if len(data['n_bedroom']) else None
    
    # extract localition and zip code
    data['location'] = soup.find('span', {"class": 'offer-details-location--locality'})
    data['location'] = data['location'].get_text() if data['location'] is not None else None
    if data['location'] is not None:
        # sous format "{Location_name} ({Zip_code})"
        data['zipcode'] = re.sub(r'(.*) \((\d+)\)', r'\g<2>', data['location'])
        data['location'] = re.sub(r'(.*) \((\d+)\)', r'\g<1>', data['location'])
   
    # then extract price and month rent
    data['price'] = soup.find('p', {"class": 'offer-price'})
    data['price'] = data['price'].get_text() if data['price'] is not None else None
    if data['price'] is not None:
        data['price'] = data['price'].replace('\n', '').replace(' ', '')
        data['price'] = int(re.sub(r'(\d+)€', r'\g<1>', data['price']))
    
    data['month_rent'] = soup.find('p', {"class": 'offer-bank'})
    data['month_rent'] = data['month_rent'].get_text() if data['month_rent'] is not None else None
    if data['month_rent'] is not None: 
        data['month_rent'] = data['month_rent'].replace('\n', '').replace(' ', '')
        data['month_rent'] = int(re.sub(r'Ou(\d+)€/mois.*', r'\g<1>', data['month_rent']))
 
    # cleaning
    data = {k: v for k, v in data.items() if v is not None}
    return data
    
```

De part ces hypothèses on peut conclure que nous allons traiter les valeurs manquantes pour n_bedroom et n_room de la façon suivante :
- On fera une interpolation linéaire entre la surface et ces deux champs (sans arrondi, pas nécessaire)
- Il y a des outliers à la distribution des données manquantes qui peuvent fortement altérer l'interpolation. 
- Avant de faire notre regression, validons l'hypothèse que les outliers (~5%) doivent être éliminés du calcul de la regression linéaire.

```python
fig, ax = plt.subplots()
df.plot(x='area', y='n_bedroom', kind='scatter', color='#8800ff', alpha=0.05, label='Bedroom', ax=ax)
df.plot(x='area', y='n_room', kind='scatter', color='#2f9599', alpha=0.05, label='Rooms', ax=ax)
plt.title('Rooms for surface and missing room number distribution')

# regression for rooms and bedrooms

x = np.array(df[df.n_bedroom.notna() & df.area.notna()].area.values)
y = np.array(df[df.n_bedroom.notna() & df.area.notna()].n_bedroom.values)
m_bedroom, b_bedroom = np.polyfit(x, y, 1)
ax.plot(x, m_bedroom * x + b_bedroom, color='#8800ff', label='Bedroom regression')  
x = np.array(df[df.n_room.notna() & df.area.notna()].area.values)
y = np.array(df[df.n_room.notna() & df.area.notna()].n_room.values)
m_room, b_room = np.polyfit(x, y, 1)
ax.plot(x, m_room * x + b_room, color='#2f9599', label='Room regression')  

# regression for rooms and bedrooms without outliers 
limit_area = round(df[df.n_bedroom.isna()].area.quantile(0.95))
x = np.array(df[(df.area < limit_area) & (df.n_bedroom == df.n_bedroom)].area.values)
y = np.array(df[(df.area < limit_area) & (df.n_bedroom == df.n_bedroom)].n_bedroom.values)
m_bedroom, b_bedroom = np.polyfit(x, y, 1)
ax.plot(x, m_bedroom * x + b_bedroom, linewidth=6, color='#ffffff')  
ax.plot(x, m_bedroom * x + b_bedroom, linewidth=3, color='#8800ff', label=f'Bedroom regression (area < {limit_area}m²)')  

limit_area = round(df[df.n_room.isna()].area.quantile(0.95))
x = np.array(df[(df.area < limit_area) & (df.n_room == df.n_room)].area.values)
y = np.array(df[(df.area < limit_area) & (df.n_room == df.n_room)].n_room.values)
m_room, b_room = np.polyfit(x, y, 1)
ax.plot(x, m_room * x + b_room, linewidth=6, color='#ffffff') 
ax.plot(x, m_room * x + b_room, linewidth=3, color='#2f9599', label=f'Room regression (area < {limit_area}m²)') 

ax.legend(loc='upper left')
ax = ax.twinx()  # instantiate a second axes that shares the same x-axis
ax.set_ylabel('Percentage of missing values in the dataset')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))

plt.hist(df[df.n_bedroom.isnull()].area, density=True, bins=30, color='#8800ff', alpha=0.15, label='Bedroom number missing')
plt.hist(df[df.n_room.isnull()].area, density=True, bins=30, color='#2f9599', alpha=0.15, label='Room number missing')
ax.legend(loc='upper right')

plt.show()
```

#### n_room et n_bedroom

Que ce soit pour n_room et n_bedroom, 
on voit très clairement la différence entre la regression linéaire sur toutes les données est bien différente de celle effectuée sur l'intervalle des données manquantes avec une plus faible surface.
Cela s'expliquement naïvement par le fait que plus un bien est grand, plus les pièces le sont, et donc, le nombre de pièce n'évolue pas de façon proportionnelle avec la surface mais légèrement logarithmiquement.

Donc notre regression sur l'intervalle de surface [0, quartile(0.95)] pour les données n_room et n_bedroom manquantes est bien plus pragmatique. Nous pouvons alors procéder à cette transformation.


```python
# change the number of rooms according to what we obtained with 0.95 quantile
print(f'Interpolation de {df.n_room.isnull().sum()} lignes où "n_bedroom" est manquant.')
print(f'Interpolation de {df.n_bedroom.isnull().sum()} lignes où "n_room" est manquant.')
df.loc[df.n_bedroom.isnull(), "n_bedroom"] = m_bedroom * df["area"] + b_bedroom
df.loc[df.n_room.isnull(), "n_room"] = m_room * df["area"] + b_room

print('Valeurs manquantes après interpolation :')
missing_values = df.isnull().mean(axis=0).reset_index(name='mean').sort_values(by='mean', ascending=False)
missing_values.plot(x='index', title='Proportion of missing values on columns', kind='bar', color='#8800ff', alpha=0.5)
missing_values[missing_values["mean"] > 0]
```

```python
# group by flag to get box plots
type_cols = ['is_appartement', 'is_maison', 'is_propriété', 'is_terrain', 'is_villa']
colors = ["#8800ff", '#2f9599', '#0088ff', '#ff8800', '#992f95']
colors = ["#8800ff", '#2f9599', '#ff8800', '#0088ff', '#333333']
alphas = [0.05, 0.05, 0.5, 0.5, 0.5]
fig, ax = plt.subplots()
for type_col, color, alpha in zip(type_cols, colors, alphas):
    name_type = type_col[3:]
    name_type = name_type[0].upper() + name_type[1:].lower()
    type_df = _df[_df[type_col] == 1]
    ax.scatter(type_df.area, type_df.price, color=color, alpha=alpha, label=name_type)
    
ax.legend()
plt.title('Distribution of the types according to the price and area')
plt.xlabel('Area (pos_log + standadized)')
plt.ylabel('Price (pos_log + standadized)')
pass
```

En effet, on remarque que c'est davantage la surface qui impact le prix avec un comportement liénaire, sans détachement visible en fonction des types de bien.
On note aussi que les appartements sont majoritairement plus petits et inversement pour les maisons et villa qui semblent avoir la même distribution surface/prix.
